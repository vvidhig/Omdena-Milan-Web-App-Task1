```python
data = data.rename(columns={'latitude': 'Latitude'})
data = data.rename(columns={'longitude': 'Longitude'})
data = data.rename(columns={'GHI (kWh/m2)': 'solar_radiation'})
data = data.rename(columns={'slope': 'Slope'})
data = data.rename(columns={'surface_roughness': 'Roughness'})
data = data.rename(columns={'ST_B10': 'LST'})
data = data.rename(columns={'soil_moisture': 'SMI'})
data = data.rename(columns={'LandUse': 'landuse'})
```
